```python
import pandas as pd
import numpy as np
```

```python
url = "https://data.cityofnewyork.us/resource/erm2-nwe9.csv?$limit=50000"
df = pd.read_csv(url)

df['created_date'] = pd.to_datetime(df['created_date'])
df = df.sort_values('created_date')
```

```python
monthly = (
    df.groupby(df['created_date'].dt.to_period('M'))
      .size()
      .reset_index(name='request_count')
)

monthly['created_date'] = monthly['created_date'].dt.to_timestamp()
monthly.head()
```

```python
monthly['month'] = monthly['created_date'].dt.month
monthly['year'] = monthly['created_date'].dt.year
```

```python
monthly['lag_1'] = monthly['request_count'].shift(1)
monthly['lag_3'] = monthly['request_count'].shift(3)
```

```python
monthly['rolling_3'] = monthly['request_count'].rolling(3).mean()
monthly['rolling_6'] = monthly['request_count'].rolling(6).mean()
```

```python
monthly.dropna().head()
```

### Feature Engineering Summary

This notebook prepares aggregated public service demand data for forecasting models.
Time-based features, lag variables, and rolling averages were created to capture
seasonality, trends, and temporal dependencies in service request volumes.
